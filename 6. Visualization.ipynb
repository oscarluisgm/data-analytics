## Global API

```python
#Here you set the size of the figure and the Title

plt.figure(figsize=(12,6))
plt.title("My Nice Plot")

#Here you used the plt.plot(x, y) method

plt.subplot(1,2,1) #subplot(row, column, panel)
plt.plot(x, x**2) #1st plot #plot(x,y)
plt.plot([0,0,0], [-10,0,100]) #2nd plot #plot([x1,x2,x3],[y1,y2,y3])... x1 with y1, x2 with y2, and x3 with y3 to create the points of the curve
plt.legend(["x2","Vertical Line"]) #legend(["1st plot"], ["2nd plot"])
plt.xlabel("X")
plt.ylabel("X Squared")

plt.subplot(1,2,2)
plt.plot(x, -1*(x**2))
plt.plot([-10,0,10], [-50,-50,-50])
plt.legend(["-x2","Horizontal Line"])
plt.xlabel("X")
plt.ylabel("X Squared")
```

## OO API

```python
# plt.subplots(): Returns an tuple containing:

# 1)-Fig:Figure Object
# 2)-Axes:Axes object or array of axes object
 

fig, axes=plt.subplots(figsize=(12,6))

# some line shapes -o, --, -., : after those comes the color initial letter g(green), c(cyan), b(blue), r(red)
                
             #marker="-o" 
axes.plot(x,x+0, "-og", label="solid green", linewidth=3)
axes.plot(x,x+1, "--c", label="dashed cyan")
axes.plot(x,x+2, "-.b", label="dasheddot blue")
axes.plot(x,x+3, ":r", label="dotted red")

axes.set_title("My Nice Plot")
axes.legend()
axes.set_xlabel("X")
axes.set_ylabel("X+N")
```

```python
N=50
##random.rand (Uniform from 0 to 1)
x=np.random.rand(N)
y=np.random.rand(N)

colors=np.random.rand(N) # array of 50 random numbers from 0 to 1 (every rgb color has a value from 0 to 1)


        #ROUND AREA= pi*R^2
area= np.pi * (20*np.random.rand(N))**2 #0 to 15 point radii
```

```python
#Histograms

values=np.random.randn(1000)
```

```python
plt.subplots(figsize=(12,6))

#histogram
plt.hist(values, bins=100, alpha=0.8
        ,histtype="bar", color="steelblue",
         edgecolor="green" 
        )

#limits of the X-axis

plt.xlim(xmin=-5,xmax=5)

#saving the image
plt.savefig("hist.png",dpi=100, bbox_inches="tight")  #To execute these commands of this cell put it before plt.show()

plt.show()

```

### KERNEL DENSITY ESTIMATION

```python
from scipy import stats

density= stats.gaussian_kde(values, bw_method=0.1) #bw_method (bandwidth) makes the graph smoother or not depending, if it's higher is more rounded if it is smaller is less rounded
density
```

```python
plt.subplots(figsize=(12,6))
                    #(min, max, points between the minimum and the maximum)
values2=np.linspace(min(values)-10, max(values)+10, 100)

#plot (not the edge but in this case seems one)

plt.plot(values2,density(values2), alpha=0.5, color="#FF7F00")

#fill
plt.fill_between(values2,0,density(values2), alpha=0.5, color="#FF7F00")

plt.xlim(xmin=-5,xmax=5)

plt.show()
```

### COMBINED PLOTS

```python
plt.subplots(figsize=(12,6))

#histogram                       #density=1 or density=True is the same, used to show the other plot 
plt.hist(values, bins=100, alpha=0.8, density=True,
         histtype="bar", color="steelblue",
         edgecolor="green" 
        )


#the other plot
plt.plot(values2,density(values2), color="#FF7F00", linewidth=3)


plt.xlim(xmin=-5,xmax=5)

plt.show()
```
